```python
from openai import AsyncOpenAI
import os

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY", "YOUR_OPENAI_API_KEY_HERE")

openai_client = AsyncOpenAI(api_key=OPENAI_API_KEY)
```

```python
openai_client
```

```python
async def synthesize_speech(text: str, voice: str = "nova") -> bytes:
        """
        Text-to-Speech using OpenAI TTS-1.
        Falls back gracefully when OPENAI_API_KEY is not configured.

        Args:
            text: Text to synthesize
            voice: alloy | echo | fable | nova | onyx | shimmer

        Returns:
            MP3 audio bytes
        """
        if not openai_client:
            raise ValueError(
                "High-quality TTS requires OPENAI_API_KEY. "
                "Browser SpeechSynthesis is available as a free alternative."
            )

        try:
            response = await openai_client.audio.speech.create(
                model="tts-1",
                voice=voice,
                input=text,
                speed=1.0
            )
            audio_bytes = response.content
            print(f"[TTS OK] {len(text)} chars -> {len(audio_bytes):,} bytes audio")
            return audio_bytes

        except Exception as exc:
            print(f"[TTS FAIL] OpenAI TTS error: {repr(exc)}")
            raise ValueError(f"Speech synthesis failed: {exc}") from exc
```

```python
synthesize_speech("Hello, this is a test of the OpenAI TTS-1 model. This text will be converted to speech using the specified voice.", voice="nova")
```

```python
audio_bytes = await synthesize_speech(
    "Hello, this is a test of the OpenAI TTS-1 model.",
    voice="nova"
)
```

```python
with open("speech.mp3", "wb") as f:
    f.write(audio_bytes)
```
